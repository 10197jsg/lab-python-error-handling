Let's enhance your code by implementing error handling to handle invalid inputs.

Follow the steps below to complete the exercise:

2. Modify the `calculate_total_price` function to include error handling.
   - If the user enters an invalid price (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the price for that product.
   - Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

3. Modify the `get_customer_orders` function to include error handling.
   - If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the number of orders.
   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

4. Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.


````python
# Step 1

products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    # Initialize inventory as a list of dictionaries containing product info (product name, quantity, price)
    inventory = []
    for product in products:
        # Get valid quantity input
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative!")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        
        # Get valid price input
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price for {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative!")
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
        
        inventory.append({"product": product, "quantity": quantity, "price": price})
    return inventory

# Step 2

def get_customer_orders(inventory):
    # Get valid number of orders input
    valid_num_orders = False
    while not valid_num_orders:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders < 1:
                raise ValueError("Number of orders must be at least 1.")
            valid_num_orders = True
        except ValueError as error:
            print(f"Error: {error}")
    
    customer_orders = []
    
    for i in range(num_orders):
        valid_order = False
        while not valid_order:
            product_name = input(f"Enter product {i+1} name: ").strip().lower()
            if any(item["product"] == product_name for item in inventory):
                product = next(item for item in inventory if item["product"] == product_name)
                if product["quantity"] > 0:
                    customer_orders.append(product_name)
                    valid_order = True
                else:
                    print(f"Sorry, {product_name} is out of stock. Please select another product.")
            else:
                print(f"Error: {product_name} is not a valid product. Please enter a valid product name.")
    
    return customer_orders

# Step 3

def calculate_total_price(customer_orders, inventory):
    total_price = 0
    for order in customer_orders:
        try:
            product = next(item for item in inventory if item["product"] == order)
            total_price += product["price"]
        except StopIteration:
            print(f"Error: {order} not found in inventory.")
    return total_price

# Step 4

def update_inventory(inventory, customer_orders):
    # Update the inventory by reducing quantity for each order
    for order in customer_orders:
        for item in inventory:
            if item["product"] == order and item["quantity"] > 0:
                item["quantity"] -= 1
    # Return updated inventory with remaining products
    return [item for item in inventory if item["quantity"] > 0]

# Step 5

def main():
    # Initialize inventory
    inventory = initialize_inventory(products)
    
    # Get customer orders
    customer_orders = get_customer_orders(inventory)
    
    # Update inventory based on orders
    inventory = update_inventory(inventory, customer_orders)
    
    # Print updated inventory
    print("Updated Inventory: ")
    for item in inventory:
        print(f"Product: {item['product']}, Quantity: {item['quantity']}, Price: {item['price']:.2f}€")
    
    # Calculate and print total price of orders
    total_price = calculate_total_price(customer_orders, inventory)
    print(f"Total Price: {total_price:.2f} €")

if __name__ == "__main__":
    main()
````
